```python
NDA = ['Bharatiya Janata Party', 'Telugu Desam', 'Janasena Party', 'Janata Dal  (United)',
      'Janata Dal  (Secular)', 'Shiv Sena', 'Lok Janshakti Party(Ram Vilas)', 'Nationalist Congress Party',
      'Sikkim Krantikari Morcha', 'United People’s Party, Liberal','Apna Dal (Soneylal)']

def classify(party):
    if party in NDA:
        return 'NDA'
    elif party == 'Independent':
        return 'No Affiliation'
    else :
        return 'INDI'
```

```python
lok_sabha_24['Alliance'] = lok_sabha_24['WinningParty'].apply(classify)
lok_sabha_24
```

```python
south = ['kerala ','karnataka ','tamil nadu ','andhra pradesh  ','telangana ']

north = ['jammu and kashmir','haryana ', 'bihar ','jharkhand ','west bengal ',
         'himachal pradesh','ladakh','punjab ','rajasthan ','uttar pradesh ','uttarakhand']

central = ['gujarat ','maharashtra ','madhya pradesh ','chhattisgarh ','odisha ']

east = ['sikkim','tripura','arunachal pradesh','assam ','manipur', 'meghalaya','mizoram','nagaland']

ut = ['nct of delhi','andaman and nicobar islands','dadra and nagar haveli and daman and diu', 'goa',
     'chandigarh','puducherry','lakshadweep']
```

```python
len(south)+len(north)+len(central)+len(east)+len(ut)
```

```python
def classify(reg):
    if reg in south:
        return 'South'
    elif reg in north:
        return 'North'
    elif reg in central:
        return 'Central'
    elif reg in east:
        return 'East'
    else:
        return 'UT'
```

```python
lok_sabha_24['Region']=lok_sabha_24['State'].apply(classify)
lok_sabha_24
```
